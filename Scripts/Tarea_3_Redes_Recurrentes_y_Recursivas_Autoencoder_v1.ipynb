## Modelo Autoencoder 1: 

El Modelo Autoencoder 1 es una red neuronal simple que comprime y reconstruye imágenes utilizando una estructura codificador-decodificador. Se utiliza para la limpieza de imágenes ruidosas.

Componentes de la arquitectura:

1. Entrada: Imágenes en escala de grises de tamaño 128x128x1.

2. Codificador:
   - Dos capas Conv2D con activación ReLU y tamaño de filtro (3, 3).
   - Dos capas MaxPooling2D para reducir la dimensionalidad.

3. Representación comprimida (codificación): Un mapa de características reducido.

4. Decodificador:

   - Dos capas Conv2D con activación ReLU y tamaño de filtro (3, 3).
   - Dos capas UpSampling2D para expandir la dimensionalidad.
   - Capa final Conv2D con activación sigmoid para reconstruir la imagen.

5. Pérdida: binary_crossentropy.
6. Optimizador: Adam.

El objetivo del modelo es aprender una representación comprimida de las imágenes ruidosas para reconstruir versiones limpias.

```python
# Modelo Autoencoder 1


import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from skimage.metrics import structural_similarity as ssim
import tensorflow as tf

# Definir rutas de las carpetas
train_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train"
clean_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train_cleaned"
test_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\test"

# Cargar y preprocesar las imágenes
def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (128, 128)) / 255.0
            images.append(img)
    return np.array(images)

train_data = load_images(train_folder)
train_cleaned_data = load_images(clean_folder)

# Expandiendo dimensiones para compatibilidad con el modelo
train_data = np.expand_dims(train_data, axis=-1)
train_cleaned_data = np.expand_dims(train_cleaned_data, axis=-1)

# Construcción del autoencoder
input_img = Input(shape=(128, 128, 1))

# Codificador
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(16, (3, 3), activation='relu', padding='same')(x)
codified = MaxPooling2D((2, 2), padding='same')(x)

# Decodificador
x = Conv2D(16, (3, 3), activation='relu', padding='same')(codified)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
output_img = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

# Modelo completo
autoencoder = Model(input_img, output_img)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# Entrenamiento
autoencoder.fit(train_data, train_cleaned_data, epochs=20, batch_size=32, validation_split=0.2)

# Evaluación visual de las imágenes
n = 5
plt.figure(figsize=(15, 5))
for i in range(n):
    # Imagen original
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(train_data[i].reshape(128, 128), cmap='gray')
    plt.title("Original")
    plt.axis("off")

    # Imagen limpia esperada
    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(train_cleaned_data[i].reshape(128, 128), cmap='gray')
    plt.title("Limpia")
    plt.axis("off")

    # Imagen reconstruida
    reconstructed = autoencoder.predict(np.expand_dims(train_data[i], axis=0))
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(reconstructed[0].reshape(128, 128), cmap='gray')
    plt.title("Reconstruida")
    plt.axis("off")
plt.show()

# Métricas de evaluación
# Función para calcular PSNR
def calculate_psnr(original, reconstructed):
    original = np.expand_dims(original, axis=-1)  # Añadir eje para canales
    reconstructed = np.expand_dims(reconstructed, axis=-1)  # Añadir eje
    return tf.image.psnr(original, reconstructed, max_val=1.0).numpy()

# Función para calcular métricas de evaluación
def evaluate_model(model, test_data, clean_data):
    mse_list, ssim_list, psnr_list = [], [], []
    for i in range(len(test_data)):
        original = clean_data[i].reshape(128, 128)
        reconstructed = model.predict(np.expand_dims(test_data[i], axis=0))[0].reshape(128, 128)

        # Calcular métricas con dimensiones correctas
        mse = np.mean((original - reconstructed) ** 2)
        ssim_value = ssim(original, reconstructed, data_range=1.0)
        psnr_value = calculate_psnr(original, reconstructed)

        mse_list.append(mse)
        ssim_list.append(ssim_value)
        psnr_list.append(psnr_value)

    # Promedio de las métricas
    return np.mean(mse_list), np.mean(ssim_list), np.mean(psnr_list)

# Evaluación del modelo Autoencoder 1
mse, ssim_value, psnr = evaluate_model(autoencoder, train_data, train_cleaned_data)
print(f"Modelo Autoencoder 1 - MSE: {mse:.4f}, SSIM: {ssim_value:.4f}, PSNR: {psnr:.2f} dB")

```

Modelo autoencoder 1

1. MSE (0.0482):

   El valor indica una diferencia moderada entre la imagen original limpia y la reconstruida. Aunque el error es pequeño, el modelo aún no reconstruye la imagen de forma óptima.

2. SSIM (0.2331):

   El bajo valor de SSIM muestra que la similitud estructural entre las imágenes es poca. La reconstrucción pierde detalles importantes en cuanto a estructura, contraste y brillo.

3. PSNR (13.25 dB):

   El PSNR bajo indica que la calidad visual reconstruida es deficiente comparada con la original. Para imágenes de alta calidad, el PSNR debería ser mayor a 20-30 dB.


Observaciones: 

* Modelo poco profundo:

   La arquitectura actual tiene dos capas convolucionales en el codificador y el decodificador. Esto puede ser insuficiente para capturar patrones complejos en imágenes.

* Tamaño de la imagen:

   Reducir la imagen a 128x128 puede llevar a pérdida de detalles importantes. Si el hardware lo permite, podríamos usar 256x256.

* Función de pérdida:

  Se está utilizando binary_crossentropy, la cual no es ideal para la reconstrucción de imágenes. Funciones como SSIM o una combinación de SSIM + MSE serían más efectivas.

* Hiperparámetros:

  El número de épocas (20) es bajo. Es posible que el modelo aún no haya convergido completamente.



## Modelo Autoencoder 2:

El Modelo Autoencoder 2 es una mejora respecto al modelo anterior.

Arquitectura:

1. Profundización de la red: 
   Capas adicionales en el codificador y decodificador.

2. LeakyReLU como función de activación: 
   Ayuda a evitar el problema de las neuronas muertas en comparación con ReLU.

3. BatchNormalization: 
   Acelera el entrenamiento y estabiliza la convergencia.

4. Dropout: Introduce regularización para reducir el sobreajuste.

5. Resolución más alta (256x256): Permite una mayor retención de detalles visuales.

Este modelo se entrena con la función de pérdida binary_crossentropy y el optimizador Adam con una tasa de aprendizaje reducida de 0.0001. El entrenamiento está monitoreado con EarlyStopping, el cual detiene el proceso si no hay mejoras en la pérdida durante 5 épocas consecutivas.

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, LeakyReLU, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from skimage.metrics import structural_similarity as ssim
import tensorflow as tf

# Definir rutas de las carpetas
train_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train"
clean_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train_cleaned"

# Cargar y preprocesar las imágenes
def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (256, 256)) / 255.0
            images.append(img)
    return np.array(images)

train_data = load_images(train_folder)
train_cleaned_data = load_images(clean_folder)

# Expandiendo dimensiones para compatibilidad con el modelo
train_data = np.expand_dims(train_data, axis=-1)
train_cleaned_data = np.expand_dims(train_cleaned_data, axis=-1)

# Construcción del autoencoder mejorado
input_img = Input(shape=(256, 256, 1))

# Codificador
x = Conv2D(32, (3, 3), padding='same')(input_img)
x = LeakyReLU(alpha=0.1)(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = BatchNormalization()(x)
x = Conv2D(64, (3, 3), padding='same')(x)
x = LeakyReLU(alpha=0.1)(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

# Decodificador
x = Conv2D(64, (3, 3), padding='same')(x)
x = LeakyReLU(alpha=0.1)(x)
x = UpSampling2D((2, 2))(x)
x = BatchNormalization()(x)
x = Conv2D(32, (3, 3), padding='same')(x)
x = LeakyReLU(alpha=0.1)(x)
x = UpSampling2D((2, 2))(x)
output_img = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

# Modelo completo
autoencoder = Model(input_img, output_img)
autoencoder.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy')

# Entrenamiento con EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
autoencoder.fit(train_data, train_cleaned_data, epochs=50, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Evaluación de métricas
def calculate_psnr(original, reconstructed):
    original = np.expand_dims(original, axis=-1)
    reconstructed = np.expand_dims(reconstructed, axis=-1)
    return tf.image.psnr(original, reconstructed, max_val=1.0).numpy()

def evaluate_model(model, test_data, clean_data):
    mse_list, ssim_list, psnr_list = [], [], []
    for i in range(len(test_data)):
        original = clean_data[i].reshape(256, 256)
        reconstructed = model.predict(np.expand_dims(test_data[i], axis=0))[0].reshape(256, 256)

        mse = np.mean((original - reconstructed) ** 2)
        ssim_value = ssim(original, reconstructed, data_range=1.0)
        psnr_value = calculate_psnr(original, reconstructed)

        mse_list.append(mse)
        ssim_list.append(ssim_value)
        psnr_list.append(psnr_value)

    return np.mean(mse_list), np.mean(ssim_list), np.mean(psnr_list)

# Resultados
mse, ssim_value, psnr = evaluate_model(autoencoder, train_data, train_cleaned_data)
print(f"Modelo Autoencoder 2 - MSE: {mse:.4f}, SSIM: {ssim_value:.4f}, PSNR: {psnr:.2f} dB")

# Visualización
n = 5
plt.figure(figsize=(15, 5))
for i in range(n):
    ax = plt.subplot(3, n, i + 1)
    plt.imshow(train_data[i].reshape(256, 256), cmap='gray')
    plt.title("Original")
    plt.axis("off")

    ax = plt.subplot(3, n, i + 1 + n)
    plt.imshow(train_cleaned_data[i].reshape(256, 256), cmap='gray')
    plt.title("Limpia")
    plt.axis("off")

    reconstructed = autoencoder.predict(np.expand_dims(train_data[i], axis=0))
    ax = plt.subplot(3, n, i + 1 + 2 * n)
    plt.imshow(reconstructed[0].reshape(256, 256), cmap='gray')
    plt.title("Reconstruida")
    plt.axis("off")
plt.show()
```

Modelo Autoencoder 2

1. MSE (Error Cuadrático Medio): 0.0494

    El valor es ligeramente mayor que el modelo anterior (0.0482), lo que indica un aumento mínimo en el error promedio entre las imágenes originales y reconstruidas.

2. SSIM (Índice de Similitud Estructural): 0.3279

    El valor es mejor que el del Modelo Autoencoder 1 (0.2331), lo cual sugiere que la estructura visual y el detalle de la imagen reconstruida se han preservado mejor.

## Modelo Autoencoder 3:

El Modelo Autoencoder 3 incorpora mejoras adicionales en la arquitectura y entrenamiento en comparación con los modelos anteriores. Se incluyen las siguientes características clave:

1. Pérdida combinada:

    * Utiliza una combinación ponderada entre SSIM (Structural Similarity Index) y MSE (Mean Squared Error), con un peso de 0.8 para SSIM y 0.2 para MSE.

    * Esta combinación favorece la preservación de la estructura de la imagen y reduce la distorsión global.

2. Skip Connections (Conexiones Residuales):

    * Se agregan conexiones entre el codificador y el decodificador para transferir información de características importantes.

    * Esto facilita una mejor reconstrucción de las imágenes al combinar características de bajo y alto nivel.

3. Codificador y decodificador más profundos:

   * Codificador:

        * 3 bloques de Conv2D, MaxPooling2D y BatchNormalization.
        * Capacidad de capturar características complejas y representaciones de mayor profundidad.

   * Decodificador:

        * 3 bloques de UpSampling2D, Conv2D, BatchNormalization y LeakyReLU.
        * Reconstrucción progresiva de la imagen utilizando las conexiones residuales.

4. Post-procesamiento para mejorar la nitidez:

    * Se implementa un filtro de nitidez basado en un kernel de convolución.
    * Mejora los bordes y detalles finos en la imagen reconstruida.


### 1. Importación de librerías y funciones auxiliares


```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, LeakyReLU, Dropout, BatchNormalization, concatenate
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from skimage.metrics import structural_similarity as ssim

# Pérdida combinada: SSIM + MSE
def combined_loss(y_true, y_pred):
    ssim_value = tf.reduce_mean(tf.image.ssim(y_true, y_pred, max_val=1.0))
    mse_value = tf.reduce_mean(tf.square(y_true - y_pred))
    return 0.8 * (1 - ssim_value) + 0.2 * mse_value

# Cargar y preprocesar imágenes
def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder, filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (256, 256)) / 255.0
            images.append(img)
    return np.array(images)

# Calcular PSNR
def calculate_psnr(original, reconstructed):
    original_exp = np.expand_dims(original, axis=(0, -1))
    reconstructed_exp = np.expand_dims(reconstructed, axis=(0, -1))
    return tf.image.psnr(original_exp, reconstructed_exp, max_val=1.0).numpy()[0]
```

### 2. Cargar datos y preparar el modelo

```python
# Rutas de las carpetas
train_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train"
clean_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\train_cleaned"
test_folder = r"C:\Entorno_Desarrollo_UDD\Machine_Learning_Avanzado\data\denoising-dirty-documents\test"

# Cargar datos
train_data = load_images(train_folder)
train_cleaned_data = load_images(clean_folder)
test_data = load_images(test_folder)

train_data = np.expand_dims(train_data, axis=-1)
train_cleaned_data = np.expand_dims(train_cleaned_data, axis=-1)
test_data = np.expand_dims(test_data, axis=-1)

# Construcción del modelo Autoencoder con skip connections
input_img = Input(shape=(256, 256, 1))

# Codificador
c1 = Conv2D(32, (3, 3), padding='same')(input_img)
c1 = LeakyReLU(alpha=0.1)(c1)
c1 = MaxPooling2D((2, 2), padding='same')(c1)
c1 = BatchNormalization()(c1)

c2 = Conv2D(64, (3, 3), padding='same')(c1)
c2 = LeakyReLU(alpha=0.1)(c2)
c2 = MaxPooling2D((2, 2), padding='same')(c2)
c2 = BatchNormalization()(c2)

c3 = Conv2D(128, (3, 3), padding='same')(c2)
c3 = LeakyReLU(alpha=0.1)(c3)
c3 = MaxPooling2D((2, 2), padding='same')(c3)
c3 = BatchNormalization()(c3)

# Decodificador
u1 = UpSampling2D((2, 2))(c3)
u1 = concatenate([u1, c2])
u1 = Conv2D(64, (3, 3), padding='same')(u1)
u1 = LeakyReLU(alpha=0.1)(u1)
u1 = BatchNormalization()(u1)

u2 = UpSampling2D((2, 2))(u1)
u2 = concatenate([u2, c1])
u2 = Conv2D(32, (3, 3), padding='same')(u2)
u2 = LeakyReLU(alpha=0.1)(u2)
u2 = BatchNormalization()(u2)

u3 = UpSampling2D((2, 2))(u2)
output = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(u3)

# Compilación del modelo
autoencoder = Model(input_img, output)
autoencoder.compile(optimizer=Adam(learning_rate=0.0001), loss=combined_loss, metrics=['mse'])
```

### 3. Entrenamiento del modelo

```python
# Entrenamiento con EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
autoencoder.fit(train_data, train_cleaned_data, epochs=75, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

### 4. Visualización de resultados

### 5. Evaluación de métricas (MSE, SSIM, PSNR)

# Métricas de Evaluación para el Proyecto de Autoencoders

## 1. **MSE (Mean Squared Error) - Error Cuadrático Medio**

El MSE mide el promedio de las diferencias cuadráticas entre los píxeles de la imagen original y los píxeles de la imagen reconstruida.

### **Fórmula**:
$
MSE = \frac{1}{N} \sum_{i=1}^N (I_{\text{original}} - I_{\text{reconstruida}})^2
$
- \( N \): Número total de píxeles en la imagen.

### **Interpretación**:
- Valores más bajos de MSE indican menor error y, por lo tanto, una mejor reconstrucción.
- Un **MSE = 0** significa que las imágenes son idénticas (sin error).

### **Desventaja**:
El MSE no capta bien las percepciones visuales (estructura o contraste), ya que se enfoca únicamente en las diferencias numéricas de píxeles.

---

## 2. **SSIM (Structural Similarity Index) - Índice de Similitud Estructural**

El SSIM mide la similitud entre dos imágenes considerando tres componentes:
- **Luminancia**: Compara el brillo.
- **Contraste**: Compara las diferencias de intensidad.
- **Estructura**: Evalúa la disposición espacial de los píxeles.

### **Fórmula**:
$
SSIM(x,y) = \frac{(2\mu_x \mu_y + C_1)(2\sigma_{xy} + C_2)}{(\mu_x^2 + \mu_y^2 + C_1)(\sigma_x^2 + \sigma_y^2 + C_2)}
$
- \( $\mu_x$ , $\mu_y$ \): Medias de las imágenes \( x \) y \( y \).
- \( $\sigma_x$, $\sigma_y$ \): Varianzas de \( x \) y \( y \).
- \( $\sigma_{xy} $\): Covarianza entre \( x \) y \( y \).
- \( C_1, C_2 \): Constantes de estabilidad.

### **Rango de valores**:
- **SSIM = 1**: Imágenes idénticas.
- **SSIM cercano a 0**: Imágenes completamente diferentes.

### **Interpretación**:
- Una puntuación alta de SSIM indica una mejor similitud estructural y visual.
- A diferencia del MSE, el SSIM es más sensible a la percepción visual humana.

---

## 3. **PSNR (Peak Signal-to-Noise Ratio) - Relación Señal-Ruido Pico**

El PSNR mide la relación entre la potencia máxima de una señal (píxeles de la imagen original) y el ruido introducido (error entre las imágenes original y reconstruida).

### **Fórmula**:
$
PSNR = 10 \cdot \log_{10} \left( \frac{\text{MAX}^2}{\text{MSE}} \right)
$
- **MAX**: Valor máximo posible de un píxel (1.0 si está normalizado, 255 si no).
- **MSE**: Error cuadrático medio.

### **Unidades**:
Se mide en decibelios (**dB**).

### **Interpretación**:
- Valores altos de PSNR indican una mejor calidad de la reconstrucción (menor ruido).
- Una PSNR de **30 dB o más** se considera buena en procesamiento de imágenes.

---

## **Resumen Comparativo**

| **Métrica** | **Qué mide**                  | **Rango**         | **Interpretación**                                   |
|-------------|-------------------------------|-------------------|----------------------------------------------------|
| **MSE**     | Error promedio de píxeles    | \[0, ∞)           | Valores más bajos indican menor error.             |
| **SSIM**    | Similitud estructural/visual | \[0, 1]           | Valores cercanos a 1 indican mejor similitud.      |
| **PSNR**    | Relación señal-ruido (calidad) | (0, ∞) dB         | Valores más altos indican mayor calidad visual.    |

---

## **Aplicación**
- **MSE**: Evalúa el error absoluto entre los píxeles.
- **SSIM**: Evalúa la percepción visual considerando estructura, contraste y brillo.
- **PSNR**: Mide la calidad de la reconstrucción en decibelios (**dB**), útil para evaluar el ruido residual.
